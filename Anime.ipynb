```python
df['Aired']
```

```python
#handle conflict formatting into Aired column
#Extract start and end dates
def extract_dates(aired):
    # Split dates if they contain 'to'
    if 'to' in aired:
        start_date, end_date = aired.split(' to ')
        return pd.to_datetime(start_date, errors='coerce'), pd.to_datetime(end_date, errors='coerce')
    else:
        # Handle single dates
        return pd.to_datetime(aired, errors='coerce'), np.nan

# Apply function to extract start and end dates
df[['Start_Aired', 'End_Aired']] = df['Aired'].apply(lambda x: pd.Series(extract_dates(x)))

# View results
print(df[['Aired', 'Start_Aired', 'End_Aired']])
```

```python
#drop the actual aired column
df.drop(columns=['Aired'], inplace=True)
```

```python
df.info()
```

```python
#fill nulls in end_Aired column
df['End_Aired'].fillna('00',inplace=True)
```

```python
df.isna().sum()
```

```python
#after handled aired formatting conflict i will fill nulls in premiered column using aired start date
df['Premiered'].fillna(df['Start_Aired'],inplace=True)
```

```python
#replace nulls in Broadcast
df['Broadcast'].fillna('unknown',inplace=True)
```

```python
#replace nulls in Genres with most friquent value
df.describe(include=object)
```

```python
df['Genres'].fillna('ActionAction, AdventureAdventure, FantasyFantasy',inplace=True)
```

# 4- Data Analysis & Visualization

```python
#1-Top 10 Anime by Popularity
top10_anime = df.nlargest(10, 'Popularity')[['English', 'Popularity']]
top10_anime
```

```python
#bar plot for Top 10 Animes
plt.figure(figsize=(10, 6))
plt.barh(top_10_anime['English'], top_10_anime['Popularity'], color='skyblue')
plt.xlabel('Popularity')
plt.ylabel('Anime (English Title)')
plt.title('Top 10 Anime by Popularity')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest popularity on top
plt.show()
```

```python
#2- Score distribution
plt.figure(figsize=(10, 6))
plt.boxplot(df['Score'], vert=False)
plt.xlabel('Score')
plt.title('Box Plot of Score Distribution')
plt.show()
```

```python
#3- TOP 7 years by total numbers
# Extract year from Start_Aired and create a new column
df['Year'] = df['Start_Aired'].dt.year

# Group by year and sum the members
yearwise_members = df.groupby('Year')['Members'].sum().reset_index()

# Get the top 7 years by total members
top_7_years = yearwise_members.nlargest(7, 'Members')
```

```python
#barplot for top 7 years
plt.figure(figsize=(10, 6))
plt.bar(top_7_years['Year'].astype(str), top_7_years['Members'], color='coral')
plt.ylabel('Total Members')
plt.title('Top 7 Years by Total Members')
plt.show()
```

```python
# 4)Top show type by numbers of anime

anime_Type= df.groupby('Type')['Synonyms'].count().reset_index()
top3_Type = anime_Type.nlargest(3,'Synonyms')
top3_Type
```

```python
#Pie Plot for top 3 Types
plt.figure(figsize=(8, 8))
plt.pie(top3_Type['Synonyms'], labels=top3_Type['Type'], autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'salmon'])
plt.title('Top 3 Types by Total Animes')
plt.show()
```

```python
#Top 5 Genres per number of animes
anime_Genres= df.groupby('Genres')['Synonyms'].count().reset_index()
top5_Type = anime_Genres.nlargest(5,'Synonyms')
top5_Type
```

```python
#bar Plot for top 5 Genres
plt.figure(figsize=(10, 6))
plt.bar(top5_Type['Genres'].astype(str), top5_Type['Synonyms'], color='lightgreen')
plt.xlabel('Genres')
plt.ylabel('Total animes')
plt.title('Top 5 Genres by Total Animes')
plt.xticks(rotation=30, ha='right')
plt.show()
```

```python
# 6- Top 5 Studios
anime_Studio= df.groupby('Studios')['Synonyms'].count().reset_index()
top5_Type = anime_Studio.nlargest(5,'Synonyms')
top5_Type
```

```python
#barplot for to 5 studios
plt.figure(figsize=(10, 6))
plt.bar(top5_Type['Studios'].astype(str), top5_Type['Synonyms'], color='lightgreen')
plt.xlabel('Studios')
plt.ylabel('Total animes')
plt.title('Top 5 Studios by Total Animes')
plt.xticks(rotation=30, ha='right')
plt.show()
```

```python
# 7)Top 4 Sources
anime_Source= df.groupby('Source')['Synonyms'].count().reset_index()
top4_Source = anime_Source.nlargest(4,'Synonyms')
top4_Source
```

```python
#Pie Plot for Top 4 sources
plt.figure(figsize=(8, 8))
plt.pie(top4_Source['Synonyms'], labels=top4_Source['Source'], autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'salmon','coral'])
plt.title('Top 4 Sources by Total Animes')
plt.show()
```
